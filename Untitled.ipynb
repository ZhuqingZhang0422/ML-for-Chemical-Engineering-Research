```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

mean = 0
standard_deviation = 1

x_values = np.arange(-5, 5, 0.1)
y_values = norm(mean, standard_deviation)

plt.plot(x_values, y_values.pdf(x_values))

mean_1 = 1
standard_deviation_1 = 1000

x_values = np.arange(-5, 5, 0.1)
y_values = norm(mean_1, standard_deviation_1)

plt.plot(x_values, y_values.pdf(x_values))
```

```python
mean = 0
standard_deviation = 1
mean_1 = 1
standard_deviation_1 = 1000

x_values = np.array([1,100,1000,10000])
y_values = norm(mean, standard_deviation)
y_values2 = norm(mean_1, standard_deviation_1)
print(y_values.pdf(x_values))
print(y_values2.pdf(x_values))
```
